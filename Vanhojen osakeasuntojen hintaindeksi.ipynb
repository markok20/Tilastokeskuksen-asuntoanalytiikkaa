```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
# Käytän Tallenna poiminta -toiminnolla tallentamaani linkkiä
# En ota mukaan kolmea ensimmäistä riviä ja sarakkeista vain [1, 2, 3, 4]

df = pd.read_excel('http://pxnet2.stat.fi/PXWeb/sq/6ef1c06a-ab7a-4d2f-acee-c4216b99ff1f', 
                   skiprows = 3, usecols = [1, 2, 3, 4])

# Nimeän huonosti nimetyn sarakkeen uudelleen
df = df.rename(columns = {'Unnamed: 1' : 'Kuukausi'})
df.head()
```

```python
# Uusimmat tiedot ovat tähdellä merkittyjä ennakkotietoja, jotka voivat vielä tarkentua
df['Kuukausi'].values
```

```python
# Poistan puuttuvia tietoja sisältävä rivit
df = df.dropna(subset = ['Kuukausi'])

# Poistan tähdet
df['Kuukausi'] = df['Kuukausi'].astype(str).str.strip('*')

# Lisään aikaleimat indeksiin. Formaatti, jossa M vuosiluvun Y ja kuukauden m välissä.
df.index = pd.to_datetime(df['Kuukausi'], format = '%YM%m')
df = df.drop('Kuukausi', axis = 1)

# Lyhyemmät nimet
df.columns = ['Koko maa', 'Pääkaupunkiseutu', 'Muu Suomi']

df.tail()
```

```python
# Laitan pääkaupunkiseudun ja muun Suomen kehityksen samaan kuvaan.

ax1 = df['Pääkaupunkiseutu'].plot(color = 'C0', title = 'Vanhojen osakeasuntojen neliöhinta: Pääkaupunkiseutu ja Muu Suomi')
ax1.set_ylabel('Pääkaupunkiseutu', color = 'C0', fontsize = 14)

ax2 = ax1.twinx()
df['Muu Suomi'].plot(ax = ax2, color = 'C1')
ax2.set_ylabel('Muu Suomi', color = 'C1', fontsize = 14)
```

```python
# Kehitys koko maassa

ax3 = df['Koko maa'].plot(color = 'C0', title = 'Vanhojen osakeasuntojen neliöhinta: Koko maa')
ax1.set_ylabel('Koko maa', color = 'C0', fontsize = 14)
```

```python
# Kehitys pääkaupunkiseudulla

ax4 = df['Pääkaupunkiseutu'].plot(color = 'C0', title = 'Vanhojen osakeasuntojen neliöhinta: Pääkaupunkiseutu')
ax1.set_ylabel('Pääkaupunkiseutu', color = 'C0', fontsize = 14)
```

```python
# Kehitys muualla Suomessa

ax4 = df['Muu Suomi'].plot(color = 'C0', title = 'Vanhojen osakeasuntojen neliöhinta: Muu Suomi')
ax1.set_ylabel('Muu Suomi', color = 'C0', fontsize = 14)
```
